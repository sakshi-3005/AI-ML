```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
np.random.seed(0)
student_scores = np.random.normal(loc=70, scale=10, size=100).astype(int)
student_scores = np.clip(student_scores, 0, 100)  # keep between 0 and 100
df = pd.DataFrame({'Score': student_scores})
df.head()
```

```python
mean_score = df['Score'].mean()
median_score = df['Score'].median()
mode_score = df['Score'].mode()[0]

print(f"Mean: {mean_score:.2f}")
print(f"Median: {median_score:.2f}")
print(f"Mode: {mode_score:.2f}")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Score'], bins=15, kde=True, color='lightgreen')
plt.axvline(mean_score, color='red', linestyle='--', label=f'Mean: {mean_score:.2f}')
plt.axvline(median_score, color='blue', linestyle='--', label=f'Median: {median_score:.2f}')
plt.axvline(mode_score, color='purple', linestyle='--', label=f'Mode: {mode_score:.2f}')
plt.title('Distribution of Student Scores')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['Score'], color='orange')
plt.title('Boxplot of Student Scores')
plt.xlabel('Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```
